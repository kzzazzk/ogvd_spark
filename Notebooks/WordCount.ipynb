```python
from pyspark import SparkContext
import re

sc = SparkContext("local[*]", "Media and Standard Deviation")

text_rdd = sc.textFile("data_ok.csv")

print(text_rdd.take(5))
```

```python
rows_rdd = text_rdd.map(lambda line: [float(x) for x in line.split(" ")])
print(rows_rdd.take(1))
```

```python
cols_rdd = rows_rdd.flatMap(lambda row: [(i, (x,x*x, 1)) for i, x in enumerate(row)])
print(cols_rdd.take(5))
```

```python
group_rdd = cols_rdd.reduceByKey(lambda a,b:(a[0]+b[0],a[1]+b[1],a[2]+b[2]))
print(group_rdd.take(1))
```

```python
import numpy as np
```

```python
mean_rdd = group_rdd.map(lambda t: (t[0],t[1][0] / t[1][2], (t[1][1] / t[1][2]) - (t[1][0] / t[1][2]), np.sqrt((t[1][1] / t[1][2]) - (t[1][0] / t[1][2]))))
print(mean_rdd.collect())
```
